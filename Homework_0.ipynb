* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


img_path = 'Data/Apollo_11_Launch.jpg'
out_path = 'Data/Apollo_11_Launch_Modified_CV2.jpg'

# read image as grayscale
orig_img = cv2.imread(img_path, 0)
if orig_img is not None:

    plt.figure(figsize=(5, 5))
    # set color map to grayscale
    plt.imshow(orig_img, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()

    # set crop parameters using numpy
    h, w = orig_img.shape
    start_row, end_row = int(h * .25), int(h * .75)
    start_col, end_col = int(w * .25), int(w * .75)
    cropped_image = orig_img[start_row:end_row, start_col:end_col]
    
    plt.figure(figsize=(5, 5))
    plt.imshow(cropped_image, cmap='gray')
    plt.title("Cropped Image")
    plt.axis('off')
    plt.show()
  
    # increase brightness of the image
    brightness_index = 70
    # add: make brighter | sub: make darker
    brighter_image = cv2.add(cropped_image, brightness_index)

    plt.figure(figsize=(5, 5))
    plt.imshow(brighter_image, cmap='gray')
    plt.title("Brighter Image")
    plt.axis('off')
    plt.show()

    # 
    (h_rot, w_rot) = brighter_image.shape
    # rotation matrix (anchor point, angle in degrees (neg: clockwise, pos: counter-clockwise), scale)
    rotation_matrix = cv2.getRotationMatrix2D((w_rot // 2, h_rot // 2), -90, 1)
    # rotate the image
    rotated_image = cv2.warpAffine(brighter_image, rotation_matrix, (w_rot, h_rot))

    # View the rotated image
    plt.figure(figsize=(5, 5))
    plt.imshow(rotated_image, cmap='gray')
    plt.title("Rotated Image")
    plt.axis('off')
    plt.show()

    # save the final image
    cv2.imwrite(out_path, rotated_image)
```
